```python
import torchvision
import torch.nn as nn
import torch
import torch.nn.functional as F
from torchvision import transforms, models, datasets
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
from torch import optim
device = 'cuda'
import cv2, glob, pandas as pd
%matplotlib inline
from glob import glob
```

```python
train_data_dir = 'archive/training_set/training_set'
test_data_dir = 'archive/test_set/test_set'
```

```python
from torch.utils.data import Dataset, DataLoader
class cats_dogs(Dataset):
    def __init__(self, folder):
        cats = glob(folder + '/cats/*.jpg')
        dogs = glob(folder + '/dogs/*.jpg')
        self.fpaths = cats + dogs
        
        from random import shuffle, seed
        seed(10)
        shuffle(self.fpaths)
        
        self.targets = [fpath.split('/')[-1].startswith('dog') for fpath in self.fpaths]

    def __len__(self):
        return len(self.fpaths)

    def __getitem__(self, ix):
        f = self.fpaths[ix]
        target = self.targets[ix]
        im = (cv2.imread(f)[:, :, ::-1])
        im = cv2.resize(im, (224, 224))
        im_tensor = torch.tensor(im / 255.).permute(2, 0, 1).float()
        target_tensor = torch.tensor([target]).float()
        return im_tensor, target_tensor
```

```python
data = cats_dogs(train_data_dir)
im, label = data[200]
```

```python
plt.imshow(im.permute(1,2,0).cpu())
print(label)
plt.show()
```

```python
def get_model():
    def conv_layer(ni, no, kernel_size, stride=1):
        return nn.Sequential(
            nn.Conv2d(ni, no, kernel_size, stride),
            nn.ReLU(),
            nn.BatchNorm2d(no),
            nn.MaxPool2d(2)
        )

    model = nn.Sequential(
        conv_layer(3, 64, 3),
        conv_layer(64, 512, 3),
        conv_layer(512, 512, 3),
        conv_layer(512, 512, 3),
        conv_layer(512, 512, 3),
        conv_layer(512, 512, 3),
        nn.Flatten(),
        nn.Linear(512, 1), # CORREO FINAL: O tamanho de entrada correto  512
        nn.Sigmoid()
    ).to(device)
    
    loss_fn = nn.BCELoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)
    return model, loss_fn, optimizer
```

```python
from torchsummary import summary
model, loss_fn, optimizer = get_model()
summary(model, input_size=(3, 224, 224))
```

```python
def get_data(train_path, test_path, batch_size=16):
    train_dataset = cats_dogs(train_path)
    val_dataset = cats_dogs(test_path)
    trn_dl = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, drop_last=True)
    val_dl = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
    return trn_dl, val_dl
```

```python
def train_batch(x, y, model, optimizer, loss_fn):
    model.train()
    prediction = model(x)
    batch_loss = loss_fn(prediction, y)
    batch_loss.backward()
    optimizer.step()
    optimizer.zero_grad()
    return batch_loss.item()
```

```python
@torch.no_grad()
def accuracy(x, y, model):
    model.eval()
    prediction = model(x)
    is_correct = (prediction > 0.5) == y
    return is_correct.cpu().numpy().tolist()
```

```python
@torch.no_grad()
def val_loss(x, y, model, loss_fn):
    model.eval()
    prediction = model(x)
    val_loss = loss_fn(prediction, y)
    return val_loss.item()
    
```

```python
trn_dl, val_dl = get_data(train_data_dir, test_data_dir)
model, loss_fn, optimizer = get_model()

train_losses, train_accuracies = [], []
val_losses, val_accuracies = [], []

for epoch in range(5):
    # <- Incio do bloco com indentao corrigida
    print(f"--- poca: {epoch+1}/5 ---")
    
    train_epoch_losses, train_epoch_accuracies = [], []
    for x, y in trn_dl:
        x, y = x.to(device), y.to(device)
        batch_loss = train_batch(x, y, model, optimizer, loss_fn)
        train_epoch_losses.append(batch_loss)
        is_correct = accuracy(x, y, model)
        train_epoch_accuracies.extend(is_correct)
        
    avg_train_loss = np.mean(train_epoch_losses)
    avg_train_acc = np.mean(train_epoch_accuracies)
    train_losses.append(avg_train_loss)
    train_accuracies.append(avg_train_acc)

    val_epoch_losses, val_epoch_accuracies = [], []
    for x, y in val_dl:
        x, y = x.to(device), y.to(device)
        val_batch_loss = val_loss(x, y, model, loss_fn)
        val_epoch_losses.append(val_batch_loss)
        is_correct = accuracy(x, y, model)
        val_epoch_accuracies.extend(is_correct)
        
    avg_val_loss = np.mean(val_epoch_losses)
    avg_val_acc = np.mean(val_epoch_accuracies)
    val_losses.append(avg_val_loss)
    val_accuracies.append(avg_val_acc)
    
    print(f"Perda de Treino: {avg_train_loss:.4f}, Acurcia de Treino: {avg_train_acc:.4f}")
    print(f"Perda de Validao: {avg_val_loss:.4f}, Acurcia de Validao: {avg_val_acc:.4f}")
    print("-" * 30)
```

```python
epochs = range(5)
plt.figure(figsize=(12, 5))
    
plt.subplot(1, 2, 1)
plt.plot(epochs, train_losses, 'bo-', label='Perda de Treino')
plt.plot(epochs, val_losses, 'ro-', label='Perda de Validao')
plt.title('Perda ao longo das pocas')
plt.xlabel('pocas')
plt.ylabel('Perda')
plt.legend()
    
plt.subplot(1, 2, 2)
plt.plot(epochs, train_accuracies, 'bo-', label='Acurcia de Treino')
plt.plot(epochs, val_accuracies, 'ro-', label='Acurcia de Validao')
plt.title('Acurcia ao longo das pocas')
plt.xlabel('pocas')
plt.ylabel('Acurcia')
plt.legend()
    
plt.show()
```
