```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
img = cv.imread('ab6761610000e5eb31f6ab67e6025de876475814')
```

```python
plt.imshow(img)
plt.show()
```

```python
rgb_image = cv.cvtColor(img, cv.COLOR_BGR2RGB) #RGB
plt.imshow(rgb_image)
plt.show()
```

```python
gray_image = cv.cvtColor(img, cv.COLOR_BGR2GRAY) #Escala de cinza
plt.imshow(gray_image, cmap = 'gray')
plt.show()
```

```python
img_gray_small = cv.resize(gray_image, (50,50)) #Diminuindo de 640x640 para 50x50
plt.imshow(img_gray_small, cmap='gray')
plt.show()
```

```python
plt.imshow(gray_image[10:340,:], cmap='gray') #Recortando do pixel 10 até os 340, no eixo X
plt.show()
```

```python
plt.imshow(gray_image[:,:300], cmap='gray') #Recortando até o pixel 300 no eixo Y
plt.show()
```
