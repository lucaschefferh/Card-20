```python
#Importação e configuração dispositivo
from torch_snippets import *
from torchvision.datasets import MNIST
from torchvision import transforms
import torch
device = 'cuda' if torch.cuda.is_available() else 'cpu'
```

```python
#Transformação que vai ser aplicada a cada imagem
img_transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize([0.5],[0.5]),
    transforms.Lambda(lambda x: x.to(device))
])
```

```python
#Carregamento do data set
trn_ds = MNIST('', transform=img_transform, train=True, download=True)
val_ds = MNIST('', transform=img_transform, train=False, download=True)
```

```python
from torch.utils.data import Dataset, DataLoader
#Criação do dataLoader
batch_size = 256

trn_dl = DataLoader(trn_ds, batch_size = batch_size, shuffle=True)
val_dl = DataLoader(val_ds, batch_size = batch_size, shuffle=False)
```

```python
import torch.nn as nn

class AutoEncoder(nn.Module):
    def __init__(self, latent_dim):
        super().__init__()
        self.latent_dim = latent_dim
        self.encoder = nn.Sequential( #Modelo sequencial
            nn.Linear(28*28,128), nn.ReLU(True), #Camada linear com função de ativação 
            nn.Linear(128,64), nn.ReLU(True), #Camada linear com função de ativação
            nn.Linear(64,latent_dim)) #Camada final
        self.decoder = nn.Sequential(
            nn.Linear(latent_dim,64),nn.ReLU(True), #Camada que expande
            nn.Linear(64,128),nn.ReLU(True), #Camada que expande de 64 para 128
            nn.Linear(128,28*28),nn.Tanh()) #Camada final com função de ativação TAnh

    #Define como os dados fluem
    def forward(self,x):
        x = x.view(len(x), -1)
        x = self.encoder(x)
        x = self.decoder(x)
        x = x.view(len(x), 1, 28, 28)
        return x
```

```python
#Função de treino que retorna o loss do treinamento
def train_batch(input,model,criterio,optimizer):
    model.train()
    optimizer.zero_grad()
    output = model(input)
    loss = criterion(output,input)
    loss.backward()
    optimizer.step()
    return loss

@torch.no_grad()
def validate_batch(input,model,criterion): #Função de validação que retorna o erro comparado ao treino
    model.eval()
    output = model(input)
    loss = criterion(output,input)
    return loss
```

```python
from fastprogress import master_bar, progress_bar
import matplotlib.pyplot as plt
import numpy as np

#Instancia o autoencoder, define o otimizador e o criterio de erro
model = AutoEncoder(3).to(device)
criterion = nn.MSELoss()
optimizer = torch.optim.AdamW(model.parameters(), lr=0.001, weight_decay=1e-5)

num_epochs = 5

all_train_losses = []
all_val_losses = []

mb = master_bar(range(num_epochs))

for epoch in mb:
    #Fase de treino
    model.train()
    epoch_train_losses = []
    for data, _ in progress_bar(trn_dl, parent=mb):
        loss = train_batch(data, model, criterion, optimizer)
        epoch_train_losses.append(loss.item())

    #Calcula o erro do treino por epoca
    avg_train_loss = np.mean(epoch_train_losses)
    all_train_losses.append(avg_train_loss)

    #Fase de validação
    model.eval()
    epoch_val_losses = []
    with torch.no_grad():
        for data, _ in progress_bar(val_dl, parent=mb):
            loss = validate_batch(data, model, criterion)
            epoch_val_losses.append(loss.item())

    #Calcula o erro da validação por epoca
    avg_val_loss = np.mean(epoch_val_losses)
    all_val_losses.append(avg_val_loss)

    mb.write(f'Epoch {epoch+1}/{num_epochs} - Train Loss: {avg_train_loss:.4f} | Val Loss: {avg_val_loss:.4f}')

#Gráfico que mostra o erro do treino e o erro de validação ao longo das epocas
plt.figure(figsize=(8, 5))
plt.plot(all_train_losses, label='Training Loss')
plt.plot(all_val_losses, label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Validation Loss Over Epochs')
plt.legend()
plt.grid(True)
plt.show()
```
